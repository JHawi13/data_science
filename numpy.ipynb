```python
import numpy as np 

zero_dimensional=np.array(6)
one_dimensional_array=np.array([2,6,7,8])
two_dimensionalarray=np.array([[4,6,7,8,9],
                              [6,7,8,8,7]])
two_dimensional_array_with_three_rows=np.array([[4,5,6,7,8],#first row
                                                [1,2,3,4,5],#second row
                                                [9,8,0,3,2]])#third row
three_dimensional_array=np.array([[[4,4,5,6,5,6], #assume three dimensions are made up of rows,columns and layers
                                   [8,6,5,4,6,6],  #the first three lines of this array are columns and rows in the first layer
                                   [9,6,5,4,4,3]],
                                  [[9,8,7,6,5,5],  #the last three lines are columns and rows in the second layer
                                   [8,5,4,3,2,2],
                                   [9,6,5,4,4,3]]])
print (np.__version__)
print("zero dimensional",zero_dimensional)
print("One dimensional array",one_dimensional_array)
print("Two dimensional array",two_dimensionalarray)
print("Three dimensional array",two_dimensional_array_with_three_rows)
print(one_dimensional_array[0]) #access the first value in a one dimensional array
print(two_dimensionalarray[1,0]) #access the first value in the second row
print(two_dimensional_array_with_three_rows[0,0])#access the first value on the third row
```

```python
import matplotlib.pyplot as plt

x=np.array([0,1,2,3,4])
y=np.array([0,1,2,3,4])

plt.plot(x,y, marker = "x")
plt.title("Title")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
x_axis =np.array([0,3,5,7])
y_axis =np.array([0,5,8,16])

plt.scatter(x_axis,y_axis)
plt.show()
```

```python
x=np.array(["apples","bananas","mango","grapes"])
y=np.array([4,15,20,2])

plt.barh(x,y,color="red")
plt.show
```

```python
x=np.array([210,90,50,10])
my_label=(["BMW","TOYOTA","JEEP","VOLVO"])
myexplode=([0.3,0,0,0])
plt.pie(x, labels=my_label,explode=myexplode,shadow=True)
plt.legend(title="cars")

plt.show()
```
